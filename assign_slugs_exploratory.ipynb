### This one tries to be precise, and get only the article

```python
def extract_article(URL):
    r = requests.get(URL).text
    soup = BeautifulSoup(r,"lxml")
    out = ''
    for line in soup.findAll("p"):
        if line.string == None:
            pass
        else:
            out += line.string
    return out
```

```python
def extract_alltext(URL):
    r = requests.get(URL).text
    soup = BeautifulSoup(r,"lxml")
    return soup.get_text()
```
